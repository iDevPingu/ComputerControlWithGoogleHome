```python
#지금까지의 것을 함수화
def preprocess(text):
    # 대문자를 소문자로 바꿈
    text = text.lower()
    # 온점부분을 띄워쓰기 해줌
    text = text.replace('.',' .')
    #띄어쓰기로 나눔
    words = text.split(' ')
    
    word_to_id = {}
    id_to_word = {}

    for word in words:
        if word not in word_to_id:
            new_id = len(word_to_id)
            word_to_id[word] = new_id
            id_to_word[new_id] = word
    
    corpus = np.array([word_to_id[w] for w in words])
    
    return corpus, word_to_id, id_to_word
```

# 분포 가설

#### I drink beer / We drink wine 
#### 즉 drink 주변에는 음료를 나타내는 단어가 올 확률이 높다
#### I guzzle beer / We guzzle wine
#### 비슷한 구조의 문장이므로 guzzle, drink 는가까운 의미의 단어일 것이다
#### 맥락 = 특정 단어를 중심에 둔 그 주변 단어
#### 맥락의 크기 = window size // 즉 window size = 2 라면 I guzzle, guzzle beer 의 두개의 맥락이 생긴다

#### 어떤 단어에 주목했을때 주변에 어떤 단어가 몇 번이 나오는지 확인
#### 이를 통계 기반 기법이라고 한다

```python
import numpy as np
text = 'You say goodbye and I say hello. But He does not say anything.'
corpus, word_to_id, id_to_word = preprocess(text)
```

```python
corpus
```

```python
# 동시 발생 행렬 만드는 함수
# 쉽게 말해 say 주변에 어떤단어가 몇 번 왔는지 분석
def create_co_matrix(corpus, vocab_size, window_size = 1):
    corpus_size = len(corpus)
    # vocab_size X vocab_size 만큼 0으로 채워진 2차원 배열 생성
    co_matrix = np.zeros((vocab_size,vocab_size),dtype = np.int32)
    
    for idx, word_id in enumerate(corpus):
        for i in range(1,window_size + 1):
            left_idx = idx - i
            right_idx = idx + i
            
            if left_idx >= 0:
                left_word_id = corpus[left_idx]
                co_matrix[word_id,left_word_id] += 1
            if right_idx < corpus_size:
                right_word_id = corpus[right_idx]
                co_matrix[word_id, right_word_id] += 1
                
    return co_matrix
```

```python
# 코사인 유사도란 두 벡터가 가리키는 방향이 얼마나 같은가를 수치로 알려줌
# 완전히 같다면 1, 완전이 반대(180도) 라면 -1 수직이라면 0
# 코사인 유사도 구하는 함수
# eps 는 혹시나 모든 벡터가 0 인것이 들어와도
# 계산 가능하도록 아주 작은 숫자를 더해준다
def cos_similarity(x,y,eps=1e-8):
    nx = x / np.sqrt(np.sum(x**2)+eps)
    ny = y / np.sqrt(np.sum(y**2)+eps)
    return np.dot(nx,ny)
```

```python
# i, you 의 유사도 구해보기
corpus, word_to_id, id_to_word = preprocess(text)
vocab_size = len(word_to_id)
C = create_co_matrix(corpus,vocab_size)

c0 = C[word_to_id['you']]
c1 = C[word_to_id['i']]
print(cos_similarity(c0,c1))
```

```python
# 어떤 단어가 검색어로 주어지면, 그 검색어와 비슷한 단어를
# 코사인 유사도 순으로 출력해주는 함수
# query = 검색어
# word_matrix = 단어 벡터 한데 모은 행렬
# top = 상위 몇 번째 까지 보여줄지
def most_similar(query, word_to_id, id_to_word, word_matrix, top =5):
    if query not in word_to_id:
        print("%s를 찾을 수 없습니다." %query)
        return
    print('\n[query] ' + query)
    query_id = word_to_id[query]
    query_vec = word_matrix[query_id]
    
    # 코사인 유사도 계산
    
    vocab_size = len(id_to_word)
    similarity = np.zeros(vocab_size)
    for i in range(vocab_size):
        similarity[i] = cos_similarity(word_matrix[i],query_vec)
        
    count = 0 
    # 코사인 유사도를 기준으로 내림차순으로 출력
    # argsort() -> 배열을 오름차순으로 정렬
    # 반환값은 배열의 인덱스
    # 여기서는 코사인 유사도가 큰 거부터 출력해야하므로 -1 곱해줌
    for i in (-1 * similarity).argsort():
        if id_to_word[i] == query:
            continue
        print(' %s: %s' % (id_to_word[i],similarity[i]))
        
        count += 1
        if count >= top:
            return
```

```python
import numpy as np
text = 'You say goodbye and I say hello. But He does not say anything.'
corpus, word_to_id, id_to_word = preprocess(text)
vocab_size = len(word_to_id)
C = create_co_matrix(corpus,vocab_size)
most_similar('you',word_to_id,id_to_word,C)
```

# 통계 기반 기법 개선

```python
# the car ... drive 와 같은 문장이 많다고 했을때 
# 현재 만든것으로는 the 가 drive 보다 car 에 더 연관성이 높다고 판단
# 해결을 위해 점별 상호정보량(PMI)라는 척도 사용
# 동시발생 횟수가 0 이면 log2 0 = -∞ 이므로 실제로는 양의 상호정보량(PPMI) 사용
# 이는 PMI 가 음수이면 0으로 취급


# verbose = 진행상황 출력 여부 결정하는 플래그
# verbose = True 인 경우 중간중간 진행 상황 알려줌
def ppmi(C, verbose = False, eps = 1e-8):
    M = np.zeros_like(C, dtype=np.float32)
    N = np.sum(C)
    S = np.sum(C,axis = 0)
    total = C.shape[0] * C.shape[1]
    cnt = 0
    
    for i in range(C.shape[0]):
        for j in range(C.shape[1]):
            pmi = np.log2(C[i,j] * N / (S[j] * S[i]) + eps)
            M[i,j] = max(0,pmi)
            
            if verbose:
                cnt += 1
                if cnt % (total//100) == 0:
                    print('%.1f%%완료' % (100 * cnt / total))
    
    return M
```

```python
import numpy as np
text = 'You say goodbye and I say hello. But He does not say anything.'
corpus, word_to_id, id_to_word = preprocess(text)
vocab_size = len(word_to_id)
C = create_co_matrix(corpus,vocab_size)
W = ppmi(C)

np.set_printoptions(precision = 3)
print("동시 발생 행렬")
print(C)
print('-'*50)
print('PPMI')
print(W)

```

```python
# 문제점 : 단어 수가 10만개가 되면 차원도 10만차원이 되는데 이는 말이 안됨
# 즉 차원감소가 필수적
# 중요한 정보는 최대한 유지하면서 줄이는 게 핵심
# 즉 축을 찾는 일을 수행하여야 함
# 특잇값분해(SVD)사용
# 특정 행렬 X 를 U S V 라는 세 행렬의 곱으로 분해
# U , V 는 직교행렬, S는 대각행렬(대각성분 이외는 모두 0)
# U 행렬 = 단어 공간 
# S 행렬 = 특잇값이 큰 순서로 나열 ( 해당 축의 중요도 )
# 이는 numpy 의 svd 메서드로 실행 가능

import numpy as np
import matplotlib.pyplot as plt

text = 'You say goodbye and I say hello. But He does not say anything.'
corpus, word_to_id, id_to_word = preprocess(text)
vocab_size = len(word_to_id)
C = create_co_matrix(corpus,vocab_size)
W = ppmi(C)

# SVD
#linalg = 선형대수의 약어
U ,S ,V = np.linalg.svd(W)
print("동시 발생 행렬")
print(C[0])
print("PPMI")
print(W[0])
print("SVD")
print(U[0])
print("2차원 벡터로 줄인 행렬")
print(U[0,:2])
```

```python
# 맥락과 타깃을 만드는 함수

def create_contexts_target(corpus, window_size = 1):
    # corpus에서 처음 원소, 마지막 원소 2개를 제외
    target = corpus[window_size : -window_size]
    contexts = []
    # 2개 제외한 값
    for idx in range(window_size, len(corpus)-window_size):
        cs = []
        for t in range(-window_size,window_size + 1):
            if t == 0:
                continue
            cs.append(corpus[idx+t])
        contexts.append(cs)
    return np.array(contexts), np.array(target)
```

```python
# SimpleCBOW 구현
import numpy as np
from common.layers import MatMul, SoftmaxWithLoss

class SimpleCBOW:
    def __init__(self, vocab_size, hidden_size):
        V,H = vocab_size, hidden_size
        
        #가중치 초기화
        W_in = 0.01 * np.random.randn(V,H).astype('f')
        W_out = 0.01 * np.random.randn(H,V).astype('f')
        
        #계층 생성
        self.in_layer0 = MatMul(W_in)
        self.in_layer1 = MatMul(W_in)
        self.out_layer = MatMul(W_out)
        self.loss_layer = SoftmaxWithLoss()
        
        #모든 가중치와 기울기를 리스트에 모음
        layers = [self.in_layer0, self.in_layer1, self.out_layer]
        self.params, self.grads = [], []
        for layer in layers:
            self.params += layer.params
            self.grads += layer.grads
            
        self.word_vecs = W_in
    def forward(self,contexts,target):
        h0 = self.in_layer0.forward(contexts[:,0])
        h1 = self.in_layer1.forward(contexts[:,1])
        h = (h0 + h1) * 0.5
        score = self.out_layer.forward(h)
        loss = self.loss_layer.forward(score,target)
        return loss
    
    def backward(self,dout = 1):
        ds = self.loss_layer.backward(dout)
        da = self.out_layer.backward(ds)
        da *= 0.5
        self.in_layer1.backward(da)
        self.in_layer0.backward(da)
        return None
```

```python
from common.trainer import Trainer
from common.optimizer import Adam
from common.util import preprocess, create_contexts_target, convert_one_hot

window_size = 1
hidden_size = 5
batch_size = 3
max_epoch = 1000

text = 'You say goodbye and I say hello.'
corpus, word_to_id, id_to_word = preprocess(text)

vocab_size = len(word_to_id)
contexts, target = create_contexts_target(corpus, window_size)
target = convert_one_hot(target, vocab_size)
contexts = convert_one_hot(contexts,vocab_size)

model = SimpleCBOW(vocab_size, hidden_size)
optimizer = Adam()
trainer = Trainer(model,optimizer)

trainer.fit(contexts,target,max_epoch, batch_size)
trainer.plot()
```

# CBOW 보충

```python
# 확률의 관점에서
# CBOW 알고리즘이 하는일 -> 맥락을 주면 타겟 단어가 출현할 확률을 출력
# 샤후 확률 P(A|B) 'B사건이 일어난 후의 A가 일어날 확률' 
# B라는 정보가 주어졌을 때 A가 일어날 확률

# 이를 활용해 음의 로그 기능도를 활용
# CBOW 가 수행하는 일은 손실 함수의 값을 가능한 작게 만드늑 것
# 이 때 얻어진 가중치 매개변수가 우리가 얻고자 하는 단어의 분산 표현
```

# skip-gram 모델

```python
# word2vec 의 2개의 모델 중 하나
# CBOW, skip-gram
# skip-gram 은 CBOW에서 다루는 맥락과 타깃을 역전시킨 모델

# CBOW 모델 -> 맥락으로 부터 타깃 추측
# skip-gram -> 타깃으로 부터 맥락 추측
# 그렇다 보니 타깃 하나가 입력층, 출력층은 맥락의 수만큼 존재

# 추론 기반 기법은 새로운 단어가 추가되면 기존의 가중치를 초기값으로 재학습가능
# 통계 기반 기법은 아예 새롭게 학습 해야함
```

# word2vec 속도 개선

```python
# Embedding 계층 도입
# Nagative Sampling 이라는 새로운 손실 함수 도입

# 어휘가 100만개 -> 원핫 표현의 벡터의 크기 100만 -> 너무 커
# 이를 해결 하기 위해 Embedding 계층 도입

# 은닉층이후의 계산 -> Negative Sampling 손실 함수 도입
```

```python
# Embedding 계층

# 결과적으로 수행하는 일은 단지 행렬의 특정 행을 추출하는 것 뿐
# 즉 가중치 매개변수로부터 단어에 해당하는 행을 추출하는 계층을 만들면 됨
```

```python
class Embedding:
    def __init__(self,W):
        self.params = [W]
        self.grads = [np.zeros_like(W)]
        self.idx = None
        
    def forward(self,idx):
        W, = self.params
        self.idx = idx
        out = W[idx]
        return out
    # W의 가중치를 꺼낸다움 dW[...] = 0 에서 모두 0 으로 덮어씀
    def backward(self,dout):
        dW, = self.grads
        dW[...] = 0
        
        for i, word_id in enumerate(self.idx):
            dW[word_id] += dout[i]
        # 혹은
        # np.add.at(dW,self.idx,dout)
        return None
```

```python
# 네거티브 샘플링 기법
# 은닉층 이후의 처리 (행렬 곱, softmax)

# 은닉층 이후의 문제점
# 어휘가 100만개, 은닉층 100층인 경우
# 은닉층의 뉴런과 가중치 행렬의 곱 -> 행렬의 크기 : 100 X 100만
# softmax 계층의 계산  -> 당연하게도 100만번 수행해야함
```
